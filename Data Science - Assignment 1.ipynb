```python
import pandas as pd
import random
import numpy as np

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)

num_records = 20
emp_ids = random.sample(range(1000, 9999), num_records)
names = [
    "Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ian", "Julia",
    "Kevin", "Laura", "Mike", "Nina", "Oscar", "Paula", "Quentin", "Rachel", "Steve", "Tina"
]
designations = [
    "Manager", "Developer", "Analyst", "Designer", "Consultant", "Lead", "Engineer", "Intern", "HR", "Support",
    "Architect", "Tester", "Director", "Specialist", "Coordinator", "Supervisor", "Administrator", "Trainer", None, None
]
years_of_experience = [random.randint(0, 20) for _ in range(num_records)]
age = [random.randint(21, 60) for _ in range(num_records)]
salary = [random.randint(7000, 15000) for _ in range(num_records)]

# Build the DataFrame
employee_data = {
    "EmpID": emp_ids,
    "Name": names,
    "Designation": designations,
    "Years_of_Experience": years_of_experience,
    "Age": age,
    "Salary": salary
}
Employee_Record = pd.DataFrame(employee_data)

# Display the dataset
Employee_Record
```
