```python
import pandas as pd
import random
import numpy as np

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)

num_records = 20
emp_ids = random.sample(range(1000, 9999), num_records)
names = [
    "Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Helen", "Ian", "Julia",
    "Kevin", "Laura", "Mike", "Nina", "Oscar", "Paula", "Quentin", "Rachel", "Steve", "Tina"
]
designations = [
    "Manager", "Developer", "Analyst", "Designer", "Consultant", "Lead", "Engineer", "Intern", "HR", "Support",
    "Architect", "Tester", "Director", "Specialist", "Coordinator", "Supervisor", "Administrator", "Trainer", None, None
]
years_of_experience = [random.randint(0, 20) for _ in range(num_records)]
age = [random.randint(21, 60) for _ in range(num_records)]
salary = [random.randint(7000, 15000) for _ in range(num_records)]

# Build the DataFrame
employee_data = {
    "EmpID": emp_ids,
    "Name": names,
    "Designation": designations,
    "Years_of_Experience": years_of_experience,
    "Age": age,
    "Salary": salary
}
Employee_Record = pd.DataFrame(employee_data)

# Display the dataset
Employee_Record
```

```python
## Student DataFrame

import pandas as pd

# First DataFrame: Student academic details
df_students = pd.DataFrame({
    'Student_ID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Specialisation': ['Data Science', 'Cyber Security', 'AI', 'Cloud Computing']
})

# Second DataFrame: Student contact details
df_contact = pd.DataFrame({
    'Student_ID': [103, 104, 105, 106],
    'Phone_Number': ['1234567890', '2345678901', '3456789012', '4567890123'],
    'Email': ['charlie@email.com', 'david@email.com', 
              'eva@email.com', 'frank@email.com']
})

print("Students DataFrame:")
print(df_students)

print("\nContact DataFrame:")
print(df_contact)
```

```python
# Inner Join
inner_join = pd.merge(df_students, df_contact, 
                      on='Student_ID', 
                      how='inner')

print("\nInner Join:")
print(inner_join)
```

```python
#Outer Join
outer_join = pd.merge(df_students, df_contact, 
                      on='Student_ID', 
                      how='outer')

print("\nOuter Join:")
print(outer_join)
```

```python
#Left Join
left_join = pd.merge(df_students, df_contact, 
                     on='Student_ID', 
                     how='left')

print("\nLeft Join:")
print(left_join)
```

```python
#Right Join 
right_join = pd.merge(df_students, df_contact, 
                      on='Student_ID', 
                      how='right')

print("\nRight Join:")
print(right_join)
```

```python
#Vertical Concatenation (row wise)
concat_vertical = pd.concat([df_students, df_contact], 
                            axis=0)

print("\nVertical Concatenation:")
print(concat_vertical)
```

```python
#Horizontal Concatenation (column wise)
concat_horizontal = pd.concat([df_students, df_contact], 
                              axis=1)

print("\nHorizontal Concatenation:")
print(concat_horizontal)
```

```python
#Create Sales Data DataFrame
import pandas as pd

# Sample Sales Data
df_sales = pd.DataFrame({
    'Order_ID': [1001, 1002, 1003, 1004, 1005, 1006, 1007, 1008],
    'Region': ['East', 'West', 'East', 'North', 'West', 'North', 'East', 'West'],
    'Product': ['Laptop', 'Laptop', 'Tablet', 'Tablet', 'Laptop', 'Tablet', 'Laptop', 'Tablet'],
    'Sales_Amount': [50000, 45000, 20000, 15000, 55000, 18000, 52000, 22000],
    'Quantity': [5, 4, 3, 2, 6, 2, 5, 3]
})

print(df_sales)
```

```python
#Pivot Table: Total Sales by Region
pivot_region = pd.pivot_table(
    df_sales,
    values='Sales_Amount',
    index='Region',
    aggfunc='sum'
)

print(pivot_region)
```

```python
#Pivot Table: Total Sales by Region and Product
pivot_region_product = pd.pivot_table(
    df_sales,
    values='Sales_Amount',
    index='Region',
    columns='Product',
    aggfunc='sum'
)

print(pivot_region_product)
```

```python
#Pivot Table with Multiple Aggregations
pivot_multi = pd.pivot_table(
    df_sales,
    values='Sales_Amount',
    index='Region',
    columns='Product',
    aggfunc=['sum', 'mean']
)

print(pivot_multi)
```

```python
#Pivot Table for Quantity Sold
pivot_quantity = pd.pivot_table(
    df_sales,
    values='Quantity',
    index='Product',
    columns='Region',
    aggfunc='sum'
)

print(pivot_quantity)
```

```python
#Adding Margins (Grand Total)
pivot_with_total = pd.pivot_table(
    df_sales,
    values='Sales_Amount',
    index='Region',
    columns='Product',
    aggfunc='sum',
    margins=True
)

print(pivot_with_total)
```
