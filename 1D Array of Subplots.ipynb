```python
import matplotlib.pyplot as plt

# Sample data
x = list(range(10))
y1 = [i for i in x]
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]

# Create 1D array of subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3)

# First subplot
axes[0].plot(x, y1)
axes[0].set_title("Linear")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")

# Second subplot
axes[1].plot(x, y2)
axes[1].set_title("Square")
axes[1].set_xlabel("X-axis")
axes[1].set_ylabel("Y-axis")

# Third subplot
axes[2].plot(x, y3)
axes[2].set_title("Cube")
axes[2].set_xlabel("X-axis")
axes[2].set_ylabel("Y-axis")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
